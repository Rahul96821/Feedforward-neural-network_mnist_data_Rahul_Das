```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.metrics import SparseCategoricalAccuracy
```

**Load and prepare the MNIST dataset**

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

**Build the model**

```python
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])
```

**Compile the model**

```python
model.compile(optimizer=Adam(),
              loss=SparseCategoricalCrossentropy(),
              metrics=[SparseCategoricalAccuracy()])
```

**Train the model**

```python
model.fit(x_train, y_train, epochs=5)
```

**Evaluate the model**

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f'\nTest accuracy: {test_acc}')
```

**Conclusion**

In this project, we built and trained a simple neural network using TensorFlow and Keras to classify handwritten digits from the MNIST dataset. The model architecture consisted of a Flatten layer to convert 2D image data into a 1D vector, followed by a Dense hidden layer with 128 neurons using the ReLU activation, and an output layer with 10 neurons using the softmax activation for multi-class classification.

After training for 5 epochs, the model achieved a training accuracy of approximately 98.6% and a test accuracy of about 97.8%, demonstrating strong generalization to unseen data. These results show that even a relatively simple feed-forward neural network can effectively recognize handwritten digits when trained on normalized image data.

This project highlights the effectiveness of neural networks for basic image classification tasks and provides a solid foundation for exploring more advanced architectures such as Convolutional Neural Networks (CNNs) to further improve accuracy and robustness.
